```python
import pandas as pd
df = pd.read_csv('chuncheon_villages.csv',encoding='cp949')
```

```python
result = df[df['노선번호'].isin(['100', '100-1','101','200','200-1','300','400','1', '2','3','4','6','7','7-1','7-2','7-3','8','8-1','9','10','10-1','11','12','13','13-1','14','15','16','16-1',
                            '17','18'])]
```

```python
df = result
```

```python
filtered_df = df[df['주소'].str.contains('석사동')]
stops_to_exclude = ['병무청앞', '춘천지방법원', '양우A', '한국토지주택공사', '수협', '남부시장', '모아엘가A', '강원대후문', '강원사대부고']

# 제외할 정류장을 포함하지 않는 행 필터링
filtered_df = filtered_df[~filtered_df['정류장명'].isin(stops_to_exclude)]
df_combined = filtered_df 
```

```python
df_combined
```

```python
df_combined = df_combined.drop_duplicates(subset=['정류장명'], keep='first')
```

```python
df_combined
```

```python
import folium

# Folium 지도 생성
map_center = [df_combined['위도'].mean(), df_combined['경도'].mean()]
mymap = folium.Map(location=map_center, zoom_start=14)
# 다른 위치 정보 추가
for index, row in df_combined.iterrows():
    folium.Marker(
        location=[row['위도'], row['경도']],
        popup=row['정류장명'],
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(mymap)

# 지도 출력
mymap
```

```python
from sklearn.cluster import KMeans
import folium

# K-means 클러스터링 수행
k = 4  # 클러스터의 수 (임의로 지정)
coordinates = df_combined[['위도', '경도']]
kmeans = KMeans(n_clusters=k, random_state=0).fit(coordinates)
centers = kmeans.cluster_centers_

# Folium 지도 생성
map_center = [df_combined['위도'].mean(), df_combined['경도'].mean()]
mymap = folium.Map(location=map_center, zoom_start=14)

# 클러스터 중심 위치 표시
for center in centers:
    folium.Marker(
        location=[center[0], center[1]],
        popup="클러스터 중심",
        icon=folium.Icon(color='black', icon='info-sign')
    ).add_to(mymap)

# 다른 위치 정보 추가
for index, row in df_combined.iterrows():
    folium.Marker(
        location=[row['위도'], row['경도']],
        popup=row['정류장명'],
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(mymap)

# 지도 출력
mymap
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# SSE(Sum of Squared Errors)를 계산하는 함수
def calculate_sse(data, k_range):
    sse = []
    for k in k_range:
        kmeans = KMeans(n_clusters=k, random_state=0).fit(data)
        sse.append(kmeans.inertia_)
    return sse

# K의 범위 설정
k_range = range(1, 11)  # 1부터 10까지의 클러스터 개수를 시도

# SSE 계산
sse = calculate_sse(coordinates, k_range)

# Elbow Method 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(k_range, sse, marker='o', linestyle='-')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.xticks(k_range)
plt.grid(True)
plt.show()
```

```python
import folium

# 주어진 좌표 배열
popular_centers = [
    [37.8596929, 127.74301898],
    [37.84998509, 127.7501887],
    [37.86600067, 127.75322273],
    [37.85386901, 127.74613547]
]

bus_centers = [
    [37.86593091, 127.75056273],
    [37.8566985, 127.745589],
    [37.86873286, 127.72965143],
    [37.860858, 127.761388]
]

# Folium 지도 생성
map_center = [sum([pc[0] for pc in popular_centers + bus_centers]) / (len(popular_centers) + len(bus_centers)),
              sum([pc[1] for pc in popular_centers + bus_centers]) / (len(popular_centers) + len(bus_centers))]
mymap = folium.Map(location=map_center, zoom_start=14)

# popular_centers 위치 표시
for i, center in enumerate(popular_centers):
    folium.Marker(
        location=center,
        popup=f"Popular Center {i+1}",
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(mymap)

# bus_centers 위치 표시
for i, center in enumerate(bus_centers):
    folium.Marker(
        location=center,
        popup=f"Bus Center {i+1}",
        icon=folium.Icon(color='red', icon='info-sign')
    ).add_to(mymap)

# 지도 출력
mymap
```
